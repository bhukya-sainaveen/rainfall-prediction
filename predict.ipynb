```python
# import all required libraries
import numpy as np
import pandas as pd
```

```python
# reading the data
data = pd.read_csv("data/austin_weather.csv")
data.head()
```

```python
# drop the unnecessary columns
data.drop(["Events", "Date", "SeaLevelPressureLowInches"], axis=1, inplace=True)
data.head()
```

```python
data.replace('T', 0.0, inplace=True)
data.head()
```

```python
data.replace('-', 0.0, inplace=True)
data.head()
```

```python
data.to_csv("data/austin_weather_final.csv")
```

```python
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X = data.drop(["PrecipitationSumInches"], axis=1)
Y = data["PrecipitationSumInches"]
```

```python
# reshaping it into 2d vector
Y = Y.values.reshape(-1, 1)
Y
```

```python
day_index = 798 # random day
days = [i for i in range(Y.size)]
```

```python
# initialize the linear regression classifier
clf = LinearRegression()
```

```python
# train the classifier
clf.fit(X, Y)
```

```python
# plot a graph
print("The precipitation trend graph:")
Y_float = Y.astype(float)
plt.scatter(days, Y_float.flatten().tolist(), color='g')
plt.scatter(days[day_index], Y[day_index], color='r')
plt.title("Precipitation Level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis = X.filter(["TempAvgF", "DewPointAvgF", "HumidityAvgPercent", "SeaLevelPressureAvgInches", "VisibilityAvgMiles", "WindAvgMPH"])
```

```python
print("The precipitation vs attributes trend graph:")
for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    x_vis = x_vis.astype(str)
    plt.scatter(days, x_vis[x_vis.columns.values[i][:100]], color = 'g')
    plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color = 'r' )
    plt.title(x_vis.columns.values[i])
plt. show()
```
